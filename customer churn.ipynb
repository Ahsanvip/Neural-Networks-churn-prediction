```python
import pandas as pd 
import matplotlib as plt 
import numpy as np
```

```python
df=pd.read_csv(r"Customer-Churn-Records.csv")
df.head()
```

```python
#dropping RowNumber CustomerID and surname as they have no impact on churn 
column_to_drop=['RowNumber','CustomerId','Surname']
df.drop(columns=column_to_drop,axis=1,inplace=True)
```

```python
df['Gender']=df['Gender'].map({'Female':0,'Male':1})
```

```python
df.dtypes
```

```python
#one hot encoding using pandas get_dummies 
target_columns=['Geography','Card Type']
df1=pd.get_dummies(data=df,columns=target_columns,dtype=int)
```

```python
df1.dtypes
```

```python
df1.head()
```

```python
#Scaling numerical values in the range of 0 to 1
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
cols_to_scale=['Balance','EstimatedSalary','Point Earned']
df1[cols_to_scale]=scaler.fit_transform(df1[cols_to_scale])
df1.head()
```

```python
df1['Age']=scaler.fit_transform(df1[['Age']])
df1.head()
```

```python
df.dtypes
```

```python
df1.dtypes
```

```python
X=df1.drop(columns='Exited',axis=1)
y=df1['Exited']
```

```python
X
```

```python
y
```

```python
#infamous train test split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
X_train.shape
```

```python
import tensorflow as tf
```

```python
#using tensorflow keras building a neural network with 19 inputs, 2 hidden and 1 output layers
from tensorflow.keras import Sequential
from tensorflow import keras
model=Sequential([
    keras.layers.Dense(19,input_dim=19,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(5,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
#compiling model
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=100)
```

```python
y_preds=model.predict(X_test).reshape(-1)
```

```python
model.evaluate(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
y_preds=np.round(y_preds)
print(classification_report(y_test, y_preds))
```
